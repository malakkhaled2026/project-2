```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Load the Data

```python
df = pd.read_csv('Procrastination_Clean.csv')
print(df.shape)
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

## 2. Descriptive Statistics

```python
# mean, median, mode for the numeric columns
print('Work Quality (last minute):')
print('  Mean  :', round(df['Work_Quality'].mean(), 2))
print('  Median:', df['Work_Quality'].median())
print('  Mode  :', df['Work_Quality'].mode()[0])

print()
print('Overall Academic Performance:')
print('  Mean  :', round(df['Performance'].mean(), 2))
print('  Median:', df['Performance'].median())
print('  Mode  :', df['Performance'].mode()[0])
```

```python
# quick look at the categorical columns
cat_cols = ['Age', 'Role', 'Frequency', 'Start_Time', 'Distractions', 'Phone_Hours', 'Affect']

for col in cat_cols:
    print(f'--- {col} ---')
    print(df[col].value_counts())
    print()
```

## 3. Reasons for Procrastinating (multi-select)

```python
# the reasons column has multiple answers separated by semicolons
# so we split them and count each reason individually
all_reasons = df['Reasons'].str.split(';').explode().str.strip()
all_reasons = all_reasons[all_reasons.str.len() > 10]
print(all_reasons.value_counts())
```

## 4. Visualizations

```python
# Figure 1 - how often do people procrastinate?
order = ['Always', 'Often', 'Sometimes', 'Rarely', 'Never']
counts = df['Frequency'].value_counts().reindex(order).fillna(0)

plt.figure(figsize=(8, 5))
plt.bar(order, counts.values, color='steelblue', edgecolor='white')
plt.title('How Often Do Respondents Procrastinate?')
plt.xlabel('Frequency')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('fig1_frequency.png', dpi=150)
plt.show()
```

```python
# Figure 2 - when do people start their tasks?
order2 = ['Once its announced', 'A few days after its announced',
          'A day before the due date', 'The day of the due date']
labels2 = ['Once Announced', 'A Few Days After', 'Day Before', 'Day Of']
counts2 = df['Start_Time'].value_counts().reindex(order2).fillna(0)

plt.figure(figsize=(8, 5))
plt.barh(labels2, counts2.values, color='steelblue', edgecolor='white')
plt.title('When Do Respondents Usually Start Tasks?')
plt.xlabel('Count')
plt.tight_layout()
plt.savefig('fig2_start_time.png', dpi=150)
plt.show()
```

```python
# Figure 3 - top reasons for procrastinating
top4 = all_reasons.value_counts().head(4)
labels3 = ['Distractions\n(Social Media)', 'Lack of\nMotivation',
           'Poor Time\nManagement', 'Feeling\nOverwhelmed']

plt.figure(figsize=(8, 5))
plt.bar(labels3, top4.values, color='steelblue', edgecolor='white')
plt.title('Main Reasons for Procrastinating')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('fig3_reasons.png', dpi=150)
plt.show()
```

```python
# Figure 4 - work quality vs performance ratings side by side
ratings = [1, 2, 3, 4, 5]
wq = df['Work_Quality'].value_counts().reindex(ratings).fillna(0)
perf = df['Performance'].value_counts().reindex(ratings).fillna(0)

x = np.arange(1, 6)
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, wq.values, width, label='Work Quality (last min)', color='steelblue')
plt.bar(x + width/2, perf.values, width, label='Overall Performance', color='lightsteelblue', edgecolor='grey')
plt.xticks(x, ['1\n(Very Low)', '2', '3', '4', '5\n(Very High)'])
plt.title('Self-Rated Work Quality vs. Academic Performance')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.savefig('fig4_ratings.png', dpi=150)
plt.show()
```

```python
# Figure 5 - does procrastination affect performance?
affect = df['Affect'].value_counts()

plt.figure(figsize=(7, 6))
plt.pie(affect.values, labels=affect.index, autopct='%1.1f%%', startangle=140)
plt.title('Does Procrastination Affect Your Performance?')
plt.tight_layout()
plt.savefig('fig5_affect.png', dpi=150)
plt.show()
```

```python
# Figure 6 - how many hours do people spend on their phone?
phone_order = ['Less than 2 hours', '2-4 hours', '4-7 hours', '8+ hours']
phone_counts = df['Phone_Hours'].value_counts().reindex(phone_order).fillna(0)

plt.figure(figsize=(7, 5))
plt.pie(phone_counts.values, labels=phone_order, autopct='%1.1f%%', startangle=90)
plt.title('Daily Phone Usage Among Respondents')
plt.tight_layout()
plt.savefig('fig6_phone.png', dpi=150)
plt.show()
```

## 5. Extra: Average Ratings by Procrastination Frequency

```python
# group by how often they procrastinate and see if ratings differ
freq_order = ['Never', 'Rarely', 'Sometimes', 'Often', 'Always']
grouped = df.groupby('Frequency')[['Work_Quality', 'Performance']].mean().reindex(freq_order)
grouped.round(2)
```

```python
x = np.arange(len(freq_order))
width = 0.35

plt.figure(figsize=(9, 5))
plt.bar(x - width/2, grouped['Work_Quality'], width, label='Work Quality', color='steelblue')
plt.bar(x + width/2, grouped['Performance'], width, label='Overall Performance', color='lightsteelblue', edgecolor='grey')
plt.xticks(x, freq_order)
plt.ylim(0, 5)
plt.ylabel('Average Rating (1-5)')
plt.title('Avg. Ratings by Procrastination Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```
